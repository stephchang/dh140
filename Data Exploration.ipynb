## Trimming the Data
Because this dataset has many columns, I wanted to trim to only the data I wanted to use.

I wanted to focus on looking at each state's total household, total households with food stamps/SNAP benefits in the past 12 months, and percentage of households that have SNAP, so I made sure to trim out other data that won't supplement that.|

```python
desired_columns = ['GEO_ID','GEO_NAME','B11001_001','DP03_0074E','DP03_0074P','geometry']
snap[desired_columns]
```

To preserve my new dataframe, I will declare it as a new variable.

```python
snap_trimmed = snap[desired_columns].copy()
snap_trimmed
```

Now I want to filter so that I only have the states whose percentage of total households that receive SNAP are above average in the United States. 

To start off, I have to find the average percentage of total households that receive SNAP.

```python
snappercent = snap_trimmed.DP03_0074P.mean()
snappercent
```

Then, I create a table based on the new query, with only states that have a higher than average percentage of total households that receive SNAP.

```python
snap_greater = snap_trimmed.query("DP03_0074P > 12.492307692307694")
snap_greater
```

Now I have all the states that are above the average percentage of total households that receive SNAP. 

## Counting Unique Values in a Column
Out of those that that have an above average percentage of total households that receive SNAP, I wanted to see if any of those states had the same percentage. 

```python
snap_greater['DP03_0074P'].value_counts()
```

From this, I was able to see that 3 states had 13.2%, 2 had 13.1, 2 had 13.7, while other states had unique percentages.

## Plotting
Since I have imported a shapefile, trimmed it, and created a series of queried subsets, I can now visualize my data.

I wanted to see visualize which states had the higher percentage of total households with SNAP, so I used the plot command.

```python
snap_greater.plot()
```

From this map, I see that more states on the East and South of the United States have a higher percentage of households that receive SNAP benefits, which is super interesting to see.
